```python
image = misc.imread("B2.jpg")
im = Image.open("B2.jpg").convert("LA").save("wow.png")
im1 = misc.imread("wow.png")
# the shape of the image
print (image.shape)

# Intialize a new array of zeroes with the same shape
binary = np.zeros((image.shape[0],image.shape[1]));

def binarize(pixel):
    if pixel[0] >= 128:
        pixel[0] = 1
    else:
        pixel[0] = 0
        
    return pixel[0]

for r in range(len(im1)): 
    for c in range(len(im1[r])): 
        binary[r][c] = binarize(im1[r][c]);
        
        
plt.imshow(binary, cmap = matplotlib.cm.Greys_r) 
plt.show()
```

```python
xy_cut = misc.imread("t3_new.png")
for r in range(len(xy_cut)): 
    count=0
    for c in range(len(xy_cut[r])): 
        count += xy_cut[r][c];
        if count > 191000 :
            for c in range(len(xy_cut[r])):
                xy_cut[r][c] = 0
```

```python
xy_cut = misc.imread("t3_new.png")
switch = "on"
for r in range(len(xy_cut)): 
    count=0
    for c in range(len(xy_cut[r])): 
        count += xy_cut[r][c];
    if count <= 190000:
        switch = "medium"
    if switch == "medium" and count > 190000:
        switch = "on"
    if count > 190000 and switch == "on":
        for c in range(len(xy_cut[r])):
            xy_cut[r][c] = 0
        switch = "off"
```

```python
plt.imshow(xy_cut, cmap = matplotlib.cm.Greys_r) 
plt.show()
```
